```python
import matplotlib.pyplot as plt
import numpy as np

# plot

thetas, gbes = np.loadtxt('try2_Cu_111-112atgb_lores.txt', delimiter = ' ', unpack = True)

#print(thetas)

plt.plot(thetas, gbes)
plt.xlabel('Tilt Angle (deg)')
plt.ylabel('GB Energy (J/m)')
plt.title('Grain Boundary Energy')
plt.show()
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(thetas * np.pi / 180, gbes)
ax.set_rticks([0.7, 0.8])
ax.set_theta_offset(np.pi/2)

ax.set_title("113-110atgb_spt25_LO_RES", va='bottom')
plt.show()
```
